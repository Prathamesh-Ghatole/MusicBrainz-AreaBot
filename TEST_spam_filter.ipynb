This notebook aims to identify spam areas in Wikidata.

```python
import pandas as pd
```

```python
# read predownloaded dataset. This is a sample of 44107 cities from wikidata with their population
df_wiki = pd.read_csv('sample_data/query_cities_population_entityid.csv')

# Sort dataframe in ascending order by population
df_wiki.sort_values(by=['population'], ascending=True, inplace=True)
df_wiki.head()
```

Cities with 0 population shouldn't exist. They are probably spam.
Even if the city is real, 0 population means the person who added the city probably took no efforts to fill in credible data, so the accuracy of the data is questionable.

```python
df_wiki[(df_wiki['population'] < '100') & (df_wiki['population'] > '0')]
```

Again, some cities claim to have 1.49 living humans, which is highly questionable. Non-integer population values are likely to be spam.
